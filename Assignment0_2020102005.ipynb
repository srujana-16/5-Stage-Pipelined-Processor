```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline


# Reading an image in OpenCV
img1 = cv2.imread('img1.jpg')
img2 = cv2.imread('img2.jpg')
img3 = cv2.imread('img3.jpg')
img4 = cv2.imread('img4.jpg')
# print(img)

# OpenCV reads the image in BGR format. 
# We can change that to RGB using this:
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)

# Displaying the image
fig = plt.figure(figsize = [10,10])
im = fig.add_subplot(2,2,1)
plt.imshow(img1)
plt.axis('off') 
im.set_title('Image 1')

im = fig.add_subplot(2,2,2)
plt.imshow(img2)
plt.axis('off') 
im.set_title('Image 2')

im= fig.add_subplot(2,2,3)
plt.imshow(img3)
plt.axis('off') 
im.set_title('Image 3')

im= fig.add_subplot(2,2,4)
plt.imshow(img4)
plt.axis('off') 
im.set_title('Image 4')
plt.show() 

# Saving an image
# cv2.imwrite('saved_image.jpg', img1)

## 4
# Grayscale Image
img_gray = cv2.imread('img4.jpg', 0) 
plt.figure(figsize = [10,10])
plt.imshow(img_gray,cmap='gray')
plt.axis('off')
plt.show()

## 5
# Printing the shape of the image
print("Rows, Columns, Channels:",img4.shape)
```

The images are loaded in BRG format which are then converted to BRG using OpenCV python cvtColor function.
